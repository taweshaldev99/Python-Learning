```python
import matplotlib.pyplot as plt

day = [1, 3, 5, 7, 9]
night = [2, 4, 6, 8, 10]

plt.scatter(day, night, color='r')
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=10)
plt.ylabel("Night", fontsize=10)
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 3, 5, 7, 9]
night = [2, 4, 6, 8, 10]
color=["r","g","b","orange","y"]
size=[200,50,80,100,190]                                 # changes size of points
plt.scatter(day,night,c=color,alpha=0.9,s=size,marker = "*",edgecolor="r")   # alpha = 0-1  for transparency
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=10,color="Red")
plt.ylabel("Night", fontsize=10,color="Red")
plt.show()
```

```python
import matplotlib.pyplot as plt

day = [1, 3, 5, 7, 9]
night = [2, 4, 6, 8, 10]
color=[10,20,30,40,50]
size=[200,50,80,100,190]                                  
plt.scatter(day,night,c=color,s=size,cmap="BrBG") 
T=plt.colorbar()
T.set_label("ColorBar")
plt.title("Scatter Plot", fontsize=20)
plt.xlabel("Day", fontsize=10,color="Red")
plt.ylabel("Night", fontsize=10,color="Red")
plt.show()
```
