```python
import matplotlib.pyplot as plt
import numpy as np
x=["C","C++","Pyhton", "Java"]
y=[45,55,90,67]
z=[20,30,50,45]

width=0.2
p=np.arange(len(x))
p1=[j+width for j in p]

plt.xlabel("Languages",fontsize=15)
plt.ylabel("Numbers",fontsize=15)
plt.title("Rankerrrs",fontsize=18)

plt.bar(p,y,width,color='g',label='Popularity')
plt.bar(p1,z,width,color='r',label='Popularity1')

plt.xticks(p+width/2,x,rotation=10)
plt.legend()
plt.show()
```
